```python
# Importaciones estándar para el taller
import os
import urllib.request
import pyspark.sql.functions as F
from pyspark.sql import SparkSession
from pyspark.storagelevel import StorageLevel # Importar StorageLevel

# Importaciones para ML
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.ml import Pipeline

# Importaciones para Gráficos
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar seaborn para gráficos más bonitos
sns.set_theme(style="whitegrid")



# 1. Detener la sesión de Spark existente (si hay una)
try:
    spark.stop()
    print("Sesión de Spark anterior detenida.")
except:
    print("No había sesión de Spark activa. Creando una nueva.")

# 2. Crear una NUEVA sesión de Spark con más memoria
#    Le damos 4GB de RAM al driver/executor.
#    Ajusta "4g" a "2g" o "3g" si tu máquina tiene poca RAM.
spark = SparkSession.builder \
    .appName("TallerSparkMLOps") \
    .master("local[*]") \
    .config("spark.driver.memory", "16g") \
    .getOrCreate()

print("Nueva sesión de Spark creada con 16GB de memoria.")
spark
```

```python
# --- 1a. Descargar los 12 archivos ---
# (Esto tomará tiempo y espacio en disco)

data_dir = "data"
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data"
os.makedirs(data_dir, exist_ok=True)

# Asumimos que los datos de Enero a Diciembre de 2024 están disponibles
# (En un caso real, ajustaríamos el rango a los datos publicados)
for i in range(1, 13): # Taller: Usamos 12 meses (Año Completo)
    month = f"{i:02d}" # Formatea el mes a 2 dígitos (ej. 1 -> "01")
    filename = f"yellow_tripdata_2024-{month}.parquet"
    filepath = os.path.join(data_dir, filename)
    
    if not os.path.exists(filepath):
        print(f"Descargando: {filename}...")
        url = f"{base_url}/{filename}"
        try:
            urllib.request.urlretrieve(url, filepath)
            print(f"Descarga de {filename} completa.")
        except Exception as e:
            print(f"Error descargando {filename}. Saltando este mes.")
    else:
        print(f"Archivo ya existe: {filename}")

print("Descarga de todos los archivos de 2024 finalizada.")




# --- 1b. Leer MÚLTIPLES archivos a la vez (Lectura Inicial) ---
#
# Leemos los 41.1M de filas, pero NO hacemos nada con ellas todavía.
# Esta es una operación "perezosa" (lazy).

print("Apuntando a todos los archivos parquet en el directorio 'data/'...")
df_2024_raw = spark.read.parquet(f"{data_dir}/yellow_tripdata_2024-*.parquet")

# Contemos el total de filas para 2024
total_filas = df_2024_raw.count()
print(f"Total de filas cargadas (Todo 2024): {total_filas:,}") # Ej: ~41 millones


# --- 1c. Limpieza y Preparación (Feature Engineering Básico) ---
#
# Re-usamos las lecciones aprendidas:
# 1. Calcular duración en minutos.
# 2. Filtrar datos "sucios".
# 3. Persistir el resultado para acelerar el resto del taller.

def limpiar_dataframe(df):
    """
    Aplica limpieza básica y feature engineering a nuestro DataFrame.
    """
    # 1. Calcular duración y filtrar viajes imposibles
    df_limpio = df.withColumn("duracion_minutos",
        (F.unix_timestamp(F.col("tpep_dropoff_datetime")) - F.unix_timestamp(F.col("tpep_pickup_datetime"))) / 60
    )
    
    # 2. Extraer componentes de fecha/hora
    df_limpio = df_limpio.withColumn("mes", F.month(F.col("tpep_pickup_datetime"))) \
                         .withColumn("dia_semana", F.dayofweek(F.col("tpep_pickup_datetime"))) \
                         .withColumn("hora_del_dia", F.hour(F.col("tpep_pickup_datetime")))

    # 3. Filtrar datos sucios
    # Viajes de más de 1 min y menos de 12 horas (720 min)
    # Distancia y costo positivos
    df_limpio = df_limpio.filter(
        (F.col("duracion_minutos") > 1) &
        (F.col("duracion_minutos") < 720) &
        (F.col("trip_distance") > 0) &
        (F.col("total_amount") > 0)
    )
    
    # 4. Calcular propina_pct
    df_limpio = df_limpio.withColumn("propina_pct",
        (F.col("tip_amount") / F.col("total_amount")) * 100
    )
    
    # 5. Filtrar propinas imposibles (ej. más de 50%) y nulos en columnas clave
    df_limpio = df_limpio.filter(
        (F.col("propina_pct").isNotNull()) &
        (F.col("propina_pct") <= 50) &
        (F.col("passenger_count").isNotNull()) &
        (F.col("RatecodeID").isNotNull())
    )
    
    return df_limpio

# --- SOLUCIÓN AL ERROR 'Java heap space' (Versión Final) ---
#
# Patrón ETL Iterativo: Procesamos un mes a la vez para
# nunca sobrecargar la memoria RAM.
#
ruta_checkpoint = "data/df_limpio_checkpoint.parquet"

# Revisa si el checkpoint ya existe para no reprocesar innecesariamente
if not os.path.exists(ruta_checkpoint):
    print("Iniciando 'Trabajo 1': Limpieza y Checkpointing Iterativo...")

    for i in range(1, 13):
        month = f"{i:02d}"
        filename = f"yellow_tripdata_2024-{month}.parquet"
        filepath = os.path.join(data_dir, filename)
        
        if os.path.exists(filepath):
            print(f"Procesando mes: {month}...")
            
            # 1. Leer UN solo mes
            df_mes = spark.read.parquet(filepath)
            
            # 2. Limpiar ESE mes
            df_mes_limpio = limpiar_dataframe(df_mes)
            
            # 3. Guardar el resultado
            if i == 1:
                # El primer mes SOBREESCRIBE
                df_mes_limpio.write.mode("overwrite").parquet(ruta_checkpoint)
            else:
                # Los meses siguientes AÑADEN
                df_mes_limpio.write.mode("append").parquet(ruta_checkpoint)
                
            print(f"Mes {month} procesado y guardado.")

    print(f"Checkpoint completado. Datos limpios guardados en: {ruta_checkpoint}")
else:
    print(f"Checkpoint ya existe en: {ruta_checkpoint}. Saltando 'Trabajo 1'.")


# --- TRABAJO 2: Leer los datos limpios ---
#
# A partir de ahora, todo el taller usará este DataFrame pre-procesado.
# La lectura será mucho más rápida.
print(f"Leyendo DataFrame limpio desde el checkpoint: {ruta_checkpoint}...")
df_limpio = spark.read.parquet(ruta_checkpoint)

# --- NO MÁS CACHE ---
# df_limpio.cache() # ELIMINAMOS ESTA LÍNEA para evitar OOM

conteo_limpio = df_limpio.count() # Verificamos el conteo leyendo del disco
print(f"DataFrame limpio cargado. Total de filas: {conteo_limpio:,}")

```

```python
# --- Pregunta 1: ¿Cómo evolucionaron los viajes y las propinas a lo largo del año? ---

# Esta agregación se ejecuta sobre ~35M+ de filas pero devuelve solo 12.
df_por_mes = df_limpio.groupBy("mes").agg(
    F.count("*").alias("total_viajes"),
    F.avg("propina_pct").alias("propina_promedio"),
    F.avg("trip_distance").alias("distancia_promedio")
).orderBy("mes")

print("--- Análisis por Mes ---")
df_por_mes.show()





# --- Pregunta 2: ¿A qué hora del día se dan las mejores propinas? ---

df_por_hora = df_limpio.groupBy("hora_del_dia").agg(
    F.count("*").alias("total_viajes"),
    F.avg("propina_pct").alias("propina_promedio")
).orderBy("propina_promedio", ascending=False)

print("--- Análisis por Hora del Día (Top 5 Horas para Propinas) ---")
df_por_hora.show(5)





# --- Pregunta 3: ¿Tarjeta (1) vs Efectivo (2)? ¿Quién da más propina? ---

# (Filtramos solo tarjeta y efectivo)
df_por_pago = df_limpio.filter(F.col("payment_type").isin(1, 2)) \
                       .withColumn("tipo_pago_nombre",
                                   F.when(F.col("payment_type") == 1, F.lit("Tarjeta"))
                                    .otherwise(F.lit("Efectivo"))
                       ) \
                       .groupBy("tipo_pago_nombre").agg(
                           F.avg("propina_pct").alias("propina_promedio"),
                           F.count("*").alias("conteo_viajes")
                       )

print("--- Análisis Tarjeta vs Efectivo ---")
df_por_pago.show()





```

```python

# --- Gráfico 1: Viajes por Mes ---

# 1. Agregamos en Spark (sobre 35M+ de filas)
df_por_mes_agg = df_limpio.groupBy("mes").count().orderBy("mes")

# 2. Convertimos a Pandas (SOLO 12 filas)
print("Convirtiendo df_por_mes_agg a Pandas...")
pd_por_mes = df_por_mes_agg.toPandas()
print(pd_por_mes)

# 3. Graficamos con Matplotlib/Seaborn
plt.figure(figsize=(10, 4))
sns.barplot(x="mes", y="count", data=pd_por_mes, palette="viridis")
plt.title("Total de Viajes por Mes (Todo 2024)")
plt.xlabel("Mes")
plt.ylabel("Número de Viajes (Millones)")
plt.show()




# --- Gráfico 2: Correlación Distancia vs Propina ---

# ¡Sería una locura pasar 35M+ de filas a Pandas para un scatter plot!
# La solución: Tomar una MUESTRA aleatoria en Spark.
df_sample = df_limpio.sample(withReplacement=False, fraction=0.001, seed=42) # 0.1% de los datos

# Ahora convertimos esta muestra (ej. ~35,000 filas) a Pandas
print("Convirtiendo muestra a Pandas...")
pd_sample = df_sample.select("trip_distance", "propina_pct").toPandas()

# Graficamos la correlación
plt.figure(figsize=(10, 6))
sns.scatterplot(x="trip_distance", y="propina_pct", data=pd_sample.query("trip_distance < 30"), alpha=0.1)
plt.title("Correlación: Distancia del Viaje vs. % de Propina (Muestra de 0.1%)")
plt.xlabel("Distancia del Viaje (millas)")
plt.ylabel("% de Propina")
plt.show()
```

```python

# --- 4a. Preparar los Features ---
# K-Means solo quiere una columna de "features".
# Usamos VectorAssembler para combinar nuestras columnas.

feature_cols = ["hora_del_dia", "dia_semana", "trip_distance", "duracion_minutos"]

# 1. Ensamblador de Vectores
assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features_unscaled"
)

# 2. Escalador Estándar
# K-Means es sensible a la escala. 'trip_distance' (ej. 0-30)
# dominaría a 'dia_semana' (ej. 1-7).
# StandardScaler centrará los datos (media 0, std 1).
scaler = StandardScaler(
    inputCol="features_unscaled",
    outputCol="features",
    withStd=True,
    withMean=True
)

# 3. El Modelo K-Means
# Busquemos 4 clústeres (ej. Hora pico, Noche, Fin de semana, Aeropuerto)
kmeans = KMeans(
    featuresCol="features",
    predictionCol="cluster_id",
    k=4,
    seed=42
)




# --- 4b. Crear el Pipeline de ML ---
# Un Pipeline encadena todos nuestros pasos.
# Esto es CRUCIAL para evitar "fugas de datos" y simplificar el código.

pipeline = Pipeline(stages=[assembler, scaler, kmeans])

print("Pipeline de ML creado. Consta de 3 etapas: Assembler, Scaler, KMeans.")




# --- 4c. Entrenar el Modelo ---
# Entrenamos el pipeline sobre nuestros datos limpios.
# Spark distribuirá el cálculo de K-Means en todo el clúster.

print("Iniciando entrenamiento del modelo K-Means (esto puede tardar)...")
# ¡Usamos el DataFrame cacheado 'df_limpio'!
model = pipeline.fit(df_limpio)
print("Entrenamiento completado.")




# --- 4d. Aplicar el Modelo (Predicción) ---
# Usamos el modelo entrenado para "etiquetar" cada uno
# de nuestros ~35M+ de viajes con un 'cluster_id' (0, 1, 2, o 3).

print("Aplicando el modelo para predecir clústeres...")
df_clustered = model.transform(df_limpio)

print("¡Predicción completa! Mostrando los nuevos clústeres:")
df_clustered.select(feature_cols + ["cluster_id"]).show(10)

```
